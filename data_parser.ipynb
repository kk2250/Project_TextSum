```python
import gensim
import os
```

```python
data_root_dir = 'data/DUC2001'
txt_opn_tag = '<TEXT>'
txt_close_tag = '</TEXT>'
```

```python
def get_text_from_doc(document_path,txt_opn_tag,txt_close_tag):
    with open(document_path, 'r') as f:
        content = f.read()
        
    start = content.index(txt_opn_tag) + len(txt_opn_tag)
    end   = content.index(txt_close_tag)
    return gensim.utils.simple_preprocess(content[start:end])
```

```python
documents = []
for filename_ in os.listdir(data_root_dir):
    documents.append(get_text_from_doc(data_root_dir+'/'+filename_,txt_opn_tag,txt_close_tag))
```

```python
model = gensim.models.Word2Vec (documents, size=150, window=10, min_count=2, workers=10)
model.train(documents,total_examples=len(documents),epochs=15)
```
